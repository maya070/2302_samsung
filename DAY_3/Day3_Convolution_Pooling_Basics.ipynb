# < Simple Convolution >

## - Import libaries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## - Prepare an Image

```python
# 3 X 3 X 1 (3X3 with 1 channel)
image = np.array([[[9], [1], [0]],
                  [[9], [9], [9]],
                  [[9], [0], [1]]], dtype=np.float32)
print('image.shape', image.shape)
print(image.reshape(3,3))

plt.imshow(image.reshape([3, 3]), cmap='Greys')
plt.show()
```

## - Prepare a Mask

```python
# Convolution Mask (2X2X1)
weight = np.array([[[1.], [0.]],
                   [[1.], [0.]]])
print('weight.shape', weight.shape)
print(weight.reshape(2,2))

plt.imshow(weight.reshape([2, 2]), cmap='Greys')
plt.show()
```

## - Apply

```python
# stride = 1, padding = False
conv2d_img = np.array([[[0], [0]],
                       [[0], [0]]], dtype=np.float32)

for i in range(2):
    for j in range(2):
        conv2d_img[i][j][0] = np.sum(image[i:i+2, j:j+2] * weight)

print(conv2d_img.shape)
print(conv2d_img.reshape([2, 2]))

plt.imshow(conv2d_img.reshape([2, 2]), cmap='Greys')
plt.show()
```

# < Simple Average/Max Pooling >

## - Import libaries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## - Prepare an Image

```python
# 3 X 3 X 1 (3X3 with 1 channel)
image = np.array([[[9], [0], [0], [0], [0], [0]],
                  [[9], [0], [0], [0], [0], [0]],
                  [[9], [2], [0], [0], [0], [0]],
                  [[9], [9], [9], [9], [9], [6]],
                  [[9], [2], [0], [0], [0], [9]],
                  [[9], [0], [0], [0], [0], [9]]], dtype=np.float32)
print('image.shape', image.shape)
print(image.reshape(6, 6))

plt.imshow(image.reshape([6, 6]), cmap='Greys')
plt.show()
```

## - Average Pooling

```python
# stride = 2, padding = False
conv2d_img = np.zeros([3, 3, 1], dtype=np.float32)

for i in range(3):
    for j in range(3):
        conv2d_img[i][j][0] = np.average(image[2*i:2*i+2, 2*j:2*j+2])

print(conv2d_img.shape)
print(conv2d_img.reshape([3, 3]))

plt.imshow(conv2d_img.reshape([3, 3]), cmap='Greys')
plt.show()
```

## - Max Pooling

```python
# stride = 2, padding = False
conv2d_img = np.zeros([3, 3, 1], dtype=np.float32)

for i in range(3):
    for j in range(3):
        conv2d_img[i][j][0] = np.max(image[2*i:2*i+2, 2*j:2*j+2])

print(conv2d_img.shape)
print(conv2d_img.reshape([3, 3]))

plt.imshow(conv2d_img.reshape([3, 3]), cmap='Greys')
plt.show()
```
